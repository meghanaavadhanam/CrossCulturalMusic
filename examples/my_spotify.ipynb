```python
artist_tracks = {}

# Iterate through the DataFrame rows
for index, row in df.iterrows():
    artist_name = row['Artist Name']
    artist_id = row['Spotify ID']
    
    # Retrieve the top tracks of the artist
    top_tracks = sp.artist_top_tracks(artist_id)
    # Extract track IDs from the top tracks and store in the dictionary
    tracks_info = [{'Track Name': track['name'], 
                    'Track ID': track['id'], 
                    'Track Link': track['external_urls']['spotify']} for track in top_tracks['tracks']]   
    artist_tracks[artist_name] = tracks_info

# Create a DataFrame from the artist_tracks dictionary
tracks_df = pd.DataFrame([(artist, track_info['Track Name'], track_info['Track ID'], track_info['Track Link']) 
                          for artist, tracks_info_list in artist_tracks.items() 
                          for track_info in tracks_info_list],
                         columns=['Artist Name', 'Track Name', 'Track ID', 'Track Link'])

# Print or use the resulting DataFrame
tracks_df
```

```python
from sklearn.cluster import KMeans

# Features for clustering
features = filled_tracks_df[['Danceability', 'Energy', 'Key', 'Loudness', 'Mode', 'Speechiness', 'Acousticness',
              'Instrumentalness', 'Liveness', 'Valence', 'Tempo']]

# Number of clusters (you can adjust this based on your preference)
num_clusters = 10

# Perform k-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
filled_tracks_df['Cluster'] = kmeans.fit_predict(features)
label = kmeans.fit_predict(features)
# Print cluster centers
print("Cluster Centers:")
print(kmeans.cluster_centers_)

# Print the clustered DataFrame
print(filled_tracks_df)
```

```python
filled_tracks_df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Getting unique labels
u_labels = np.unique(filled_tracks_df['Cluster'])

# Plotting the results
for i in u_labels:
    cluster_data = filled_tracks_df[filled_tracks_df['Cluster'] == i]
    plt.scatter(cluster_data['Energy'], cluster_data['Tempo'], label=f'Cluster {i}')

plt.legend()
plt.xlabel('Energy')
plt.ylabel('Tempo')
plt.title('K-means Clustering')
plt.show()
```
